```python
db_path = r'job_applications.db'
conn = sqlite3.connect(db_path)
                                                              
query = '''
SELECT *
FROM job_applications app
left join job_descriptions desc on (app.application_id = desc.description_id)
'''

pd.read_sql(query, conn)

# conn.close()
```
